## Background and Prior Work

As one of the most famous video-sharing platforms in the world, Youtube allows users to view, upload, like/dislike, and comment on their videos. In addition, Youtube also provides a trending section to highlight videos that attract plenty of audiences’ attention. 

According to the Youtube Help page, "Trending is not personalized and displays the same list of trending videos in each country to all users." This is why one may see videos in trending in a foreign language because these videos in trending are popular worldwide. Youtube utilizes a particular algorithm to calculate the number of views and many factors for each video and then determines whether the video is popular or not. If so, the video will appear in Trending. By using this method, Trending helps users see what are the most popular videos on YouTube and the world.

The dataset we plan to analyze helps us explore our curiosities of how many youtube videos spread over the world. This dataset includes information about  trending youtube videos over 10 different regions for several months. Link to the dataset: https://www.kaggle.com/datasnaek/youtube-new


This dataset records the top trending YouTube videos in different regions on a daily basis. The dataset contains 20 files for 10 regions: the U.S, Canada, Germany, France, Great Britain, Korea, Japan, Mexico, India, and Russia. Data from the same regions are grouped into a CSV file, with each entry divided into 16 columns: “trending date”, “title”, “channel title”, “category id”, “publish time”, “tags”, “views”, “likes”, “dislikes”, “comment count”, “thumbnail link”, “comments disabled”, “ratings disabled”, “video error or removed”, and “description and rating”.

The specific name of what the “category_id” refers to is stored in a relatively small file JSON file. The JSON files are also identified by which region the data is coming from. ### 3.1 What are the relationships of each region on trending video? 

Some videos become trending in different regions at different times. Some regions are also more likely the first ones featuring popular videos in their trending section. 

Hypothesis: United States Viewer are the ones who pick up the trend fastest, while Russia Viewers are the ones who pick up the trend slowest.


```python
def set_in_region1_and_region2(region1, region2):
    return set(data[data.region == region1].title).intersection(set(data[data.region == region2].title))

def set_check(x,lst):
    if x in lst:
        return True
    return False

def all_table_region1_and_region2(region1, region2):
    return data[(data.region == region1) | (data.region == region2)]

def find_relavent_table(region1, region2):
    temp = set_in_region1_and_region2(region1,region2)
    all_table = all_table_region1_and_region2(region1, region2)
    condition = all_table.title.apply(lambda x: set_check(x,temp))
    return all_table[condition]
```

Here are the top 15 unique region pair, which have the most count of trending videos they have shared.

```python
result = dict()
for i in ['US', 'CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU']:
    for j in ['US', 'CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU']:
        if i != j and (j,i) not in result.keys():
            t = set_in_region1_and_region2(i,j)
            result[(i,j)] = len(t)
pd.DataFrame.from_dict(result,orient = 'index',columns = ['count'])\
.sort_values('count',ascending = False).iloc[:15]
```

```python
def trending_first_in(df):
    return df.sort_values(['trending_date','likes'], ascending = [True,True]).region.iloc[0]
def come_from_value_counts(df):
    return df.groupby('title').apply(trending_first_in).value_counts()
```

```python
keysl = list(result.keys())
G = nx.DiGraph()
G.add_nodes_from(data.region.unique())
edges = []
for k in keysl:
    temp = find_relavent_table(k[0],k[1])
    find_order = come_from_value_counts(temp)
    if len(find_order.index) == 2:
        edges.append((find_order.index[0],find_order.index[1]))
    else:
        lst = [k[0],k[1]]
        to = list(set(lst) - set(find_order.index[0]))
        edges.append((find_order.index[0],to[0]))
G.add_edges_from(edges)
```

```python
nx.draw_circular(G, with_labels =True, node_size = 2000,arrowsize = 15, width = 1, dpi = 1000)
```

We first did a general matched-pair analysis by comparing trending time differences (“trending_date” and number of “likes” in the dataset) of trending videos between each pair of regions to determine who catches the trend earlier. If region A catches trends faster than region B (if region A have more videos that trends earlier than region B), we connect two regions(vertices) with a directed edge (A -> B).

(The graph above is a full directed graph with 10 vertices, each representing a region, and each arrow means the trend is first caught by the source vertex and later caught by the destination vertex(Region) according to the trending time. For example, an arrow from US to CA represents that more videos trended in the US first than Canada.) As we can see from the graph above, the US is the fastest one catching the trending videos relatively; since Mexico is the sink of the graph, we infer it as the slowest catcher for trends, even slower than Russia, which is different from our hypothesis.

### 3.2 Which region is the most separated from the world?

Due to language barriers, we think Russia, Japan, Korea are more likely to have their own trending videos.

We first identified the videos that were trending in exactly one region. For each region, we calculated how many of the trending videos stayed inside and never got to other countries. We assume this number would represent how seperated a region is from the world.

```python
# Select the videos that were trending in only one region
df1 = data.drop_duplicates(subset ="title", keep = False) 
uniq = df1.region.value_counts()
df_all = data.merge(df1.drop_duplicates(), on=['title'], how='left', indicator=True)
df2 = df_all[df_all._merge != 'both']
uniq2 = df2.region_x.value_counts()
all_c = uniq + uniq2
uniq_perc = uniq / all_c
dp_perc = uniq2 / all_c
```

```python
N = 10
ind = np.arange(10)  # the x locations for the groups
width = 0.4       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot()

rects1 = ax.bar(ind, uniq_perc, width, color='deepskyblue')
rects2 = ax.bar(ind+width, dp_perc, width, color='sandybrown')

ax.set_ylabel('Percentage')
ax.set_xticks(ind+width/2)
ax.set_xticklabels( dp_perc.index )
ax.legend( (rects1[0], rects2[0] ), ('unique', 'duplicate') )
```

The data fits our hypothesis: Russia, Japan, Korea and India have their own culture circle and their trending videos rarely get out of that circle; the US and Great Britain speaks the universal language - English, thus share more trending videos with other regions.

```python
def two_sample_ratio_mean_test(region1, region2):
    return stats.ttest_ind(region1,region2)
```

```python
asia_dict = dict()
for i in asia:
    for j in asia:
        if i != j and (j,i) not in asia_dict.keys():
            i_ls = list(df1[df1.region == i].ratio)
            j_ls = list(df1[df1.region == j].ratio)
            t,p = two_sample_ratio_mean_test(i_ls, j_ls)
            asia_dict[(i,j)] = [t,p]
```

```python
temp = pd.DataFrame.from_dict(asia_dict,orient = 'index', columns = ['t statistic', 'p value'])
temp['result'] = temp['p value'].apply(lambda x: 'Similar Mean' if x > 0.05 else 'Different Mean')
temp
```
